### Training a Linear Regression Model

```python
X = aq_data[['Site Num','County Code','State Code','Parameter Code']]
y = aq_data['Sample Measurement']
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
```

### Creating and Training the Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
# print the intercept
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Sample Measurement'])
coeff_df
```

### Predictions from our Model

```python
predictions = lm.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': predictions})
df
```

```python
plt.scatter(y_test,predictions)
```

```python
# extract day of week
import datetime
los_angeles_aq['Weekday'] = pd.to_datetime(los_angeles_aq['Date Local']).dt.day_name()
```

```python
# Print aq changes throughout the day from 0:00 - 24:00. Break into seperate plots by weekday
nrows = 4
ncols = 2
plt.subplots(nrows, ncols, figsize=(15,15))
weekdays = [ 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday' ]
idx = 1
values_by_weekday = {}
for day in weekdays:
    plt.subplot(nrows, ncols, idx)
    # for loop 0:00 - 24:00
    # sum measurements where weekday == and time ==
    times = []
    for hour in range(0, 23):
        if hour <= 9:
          time = "0" + str(hour) + ":00"
        else:
          time = str(hour) + ":00"
        # select subset matching given weekday and hour
        subset = los_angeles_aq.loc[ (los_angeles_aq['Weekday'] == day) & (los_angeles_aq['Time Local'] == time) ]['Sample Measurement']
           #print(subset.mean())
        times.append(subset.min())
    values_by_weekday[day] = times
    plt.plot( times )
    plt.title(day)
    idx = idx + 1
   
```
